#**Z Test**

```python
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loc, scale = 100, 2
x = np.linspace(loc - 3*scale, loc+3*scale, 1001)
plt.plot(x, norm.pdf(x, loc = loc, scale = scale))
plt.title(f"Normal distribution mu = {loc}, sigma = {scale}")
plt.show()
```

**samping**

```python
sample_size = [9, 16, 100, 1024]
trials = len(sample_size)
exp = [0]*trials
N = 1000

for i in range(trials):
    exp[i] = np.random.randn(N, sample_size[i])*scale + loc
    exp[i] = np.mean(exp[i], axis = 1)

plt.figure(figsize = (8, 8))
for i in range(trials):
    plt.subplot(2, 2, i+1)
    sns.kdeplot(x = exp[i])
    plt.title(f"sample size = {sample_size[i]}")
    plt.text(x = min(exp[i]), y = 0.1 + ((i*0.2)), s = f"experiment sigma = {exp[i].std()}")
plt.show()

```

**lets choose n = 100: =>**
 - E(mu) = 100
 - E(std) = std_population / sqrt(n) = 0.2

```python
n = 100
mu = loc
std_sample = scale / np.sqrt(n)
x1 = np.linspace(-3*std_sample + mu, 3*std_sample + mu, 1001)
exp1 = norm.pdf(x1, loc = mu, scale = std_sample)
plt.plot(x1, exp1)
plt.title(f"Distribution of mean of trial with {n} samples")
plt.axvline(mu + 2*std_sample)
plt.axvline(mu - 2*std_sample)
plt.axvline(mu + 2.2*std_sample, color = "red")
plt.axvline(mu - 1.2*std_sample, color = "green")
plt.show()
```

**N(mu = 0, std = 1)**

```python
mean, std = 0, 1
alpha = 0.05

#z_statistic, z_critical
z_critical_1 = norm.ppf(alpha/2, loc = mean, scale = std)
z_critical_2 = norm.ppf(1 - alpha/2)

measurement = 100.3 #example
z_statistic = (measurement - loc)/std_sample

x2 = np.linspace(-3*std + mean, mean + 3*std, 1001)
exp2 = norm.pdf(x2, loc = mean, scale = std)
plt.plot(x2, exp2)
plt.title("Normal distribution")


plt.axvline(z_statistic, color = "blue")
plt.axvline(z_critical_1, color = "red")
plt.axvline(z_critical_2, color = "red")
plt.show()
```

## **p Value** for example measurement

```python
p = norm.sf(abs(z_statistic))*2
print("p = ", p*100, "%")
```

**We fail to reject the null hypothesis, p > alpha**

p value is the propability of seeing results as extreme or more extreme than result(measurement) that we got

```python
# alpha = 0.5 => false alarm rate = 0.5
from statsmodels.stats.weightstats import ztest
N = 1000
alpha = 0.05
false_alarms = [0]*N
for i in range(N):
    x1 = np.random.randn(100)*3 + 100
    x2 = np.random.randn(100)*3 + 100
    z_statistic, p = ztest(x1, x2)
    false_alarms[i] = p < alpha
print(sum(false_alarms)/N)

```

### T test VS Z test

```python
from scipy.stats import t
dof = np.arange(1, 4)**3

plt.figure(figsize = (5, 5))
plt.title("Normal Vs T")
x = np.linspace(-4, 4, 1001)
plt.plot(x, norm.pdf(x))

legend = [f"dof = {dof[i]}" for i in range(len(dof))]
legend.insert(0, "normal")
for i in range(len(dof)):
    plt.plot(x, t.pdf(x, dof[i]))
plt.legend(legend)
plt.show()
```
